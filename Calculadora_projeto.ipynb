```python
# Inserir números para a calculadora

num1 = input("Digite o primeiro número: ")
print('Primeiro número:', num1)
num2 = input("Digite o segundo número: ")
print('Segundo número:', num2)
```

```python
# Converte os valores recebidos em int e confirma a conversão

num1_convert = int(num1)
print(type(num1_convert))
num2_convert = int(num2)
print(type(num2_convert))

```

```python
# Estabelece as operações possíveis

soma = num1_convert + num2_convert
print(soma)
subt = num1_convert - num2_convert
print(subt)
mult= num1_convert * num2_convert
print(mult)
divi = num1_convert / num2_convert
print(divi)
restdiv = num1_convert % num2_convert
print(restdiv)
```

```python
# Pergunta qual operação o usuário quer e traz o resultado, recomeça a operação se for usado um caracter inválido

result= input('Digite a operação desejada: + para soma, - para subtração, * para multiplicação, / para divisão, % para resto da divisão: ')
if result == '+':
  print("A soma dos elementos é", soma)
elif result == '-':
  print('A subtração dos elementos é',subt)
elif result == "*":
  print('A multiplicação dos elementos é', mult)
elif result == "/":
  print("A divisão dos elementos é", divi)
elif result == "%":
  print("O resto da divisão dos elementos é", restdiv)
else:
  print("Erro, por favor selecione uma operação válida.")

```
